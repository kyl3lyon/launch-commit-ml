```python
import pandas as pd
import json
```

**Data**

```python
launch_sites_df = pd.read_json("data/launch_sites.json")
selected_launch_sites_df = pd.read_json("data/selected_launch_sites.json")
```

```python
unique_launch_sites = launch_sites_df["code"].unique()
unique_selected_launch_sites = selected_launch_sites_df["SITE_CODE"].unique()
```

```python
def analyze_launch_sites(unique_launch_sites, unique_selected_launch_sites):
    # Convert arrays to sets for efficient comparison
    launch_sites_set = set(unique_launch_sites)
    selected_sites_set = set(unique_selected_launch_sites)

    # Find matching sites
    matching_sites = launch_sites_set.intersection(selected_sites_set)

    # Find sites in selected_launch_sites that don't match
    unmatched_selected_sites = selected_sites_set - launch_sites_set

    # Find sites in launch_sites that don't match
    unmatched_launch_sites = launch_sites_set - selected_sites_set

    # Count of matches
    match_count = len(matching_sites)

    # Generate result string
    result = f"Number of matching launch sites: {match_count}\n\n"

    result += "Matching launch sites:\n"
    for site in matching_sites:
        result += f"- {site}\n"

    result += "\nSelected launch sites that don't match:\n"
    for site in unmatched_selected_sites:
        result += f"- {site}\n"

    result += "\nLaunch sites that are not in the selected list:\n"
    for site in unmatched_launch_sites:
        result += f"- {site}\n"

    return result

# Usage
result = analyze_launch_sites(unique_launch_sites, unique_selected_launch_sites)
print(result)
```
